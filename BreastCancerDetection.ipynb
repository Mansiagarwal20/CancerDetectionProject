## Splitting the Dataset into Training , Testing and Validation

```python
X_train = feature[:500]
Y_train = label[:500]
X_val = feature[500:535]
Y_val = label[500:535]
X_test = feature[535:]
Y_test = label[535:]
```

## Checking if the Splitting is done is correct proportion

```python
print(X_train.shape)
print(Y_train.shape)
print(X_val.shape)
print(Y_val.shape)
print(X_test.shape)
print(Y_test.shape)
```

## **Deep Learning Model Implementation**

### Creating an Empty Neural Network 

```python
model = Sequential()
```

### Adding the Layers of the Neural Network & Compiling the Model

```python
model.add(Dense(15 , activation= 'relu', input_dim = 30))
model.add(Dense(8 , activation= 'relu'))
model.add(Dense(1 , activation= 'sigmoid'))
model.compile(loss ='binary_crossentropy' , optimizer= 'adam' , metrics= ['accuracy'])
```

### Training the Model

```python
model.fit(X_train , Y_train , epochs= 10, validation_data = (X_val , Y_val)) 
```

```python
model.evaluate(X_test,Y_test)
```

### Passing One Tumor at a Time through the Neural Network

```python
model.fit(X_train , Y_train , batch_size = 1 , epochs=10 , validation_data = (X_val , Y_val)) 
```

```python
model.evaluate(X_test,Y_test)
```

### Predicting the Nature of Tumor Using the Model

```python
for i in range(len(X_test)):
  sample = X_test[i]
  sample = np.reshape(sample,(1,30))
  if (model.predict(sample) > 0.5):
    print("By Model Prediction the Tumor is - Benign")
  else:
    print("By Model Prediction the Tumor is - Malignant")
  if (Y_test[i]==0):
    print("Actual Nature of Tumor is - Malignant")
  else:
      print("Actual Nature of Tumor is - Benign")
  print("-------------\n")
```
